### Define functions to plot visualizations

```python
def plot_heatmap(title, corr, figsize = (12,12),cmap="RdBu_r"):
    fig = plt.figure(figsize=figsize)
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True
    ax = sns.heatmap(corr, mask=mask, annot=True,cmap=cmap)
    ax.set_title(f'Correlation Heatmap: {title}')

# plt.savefig('../plots/corr_heatmap.png');
```

```python
def plot_box(df, list_of_columns, y, figsize = (6,4)):
    for col in list_of_columns:
        fig = plt.figure(figsize=figsize)
        ax = sns.boxplot(x=df[col], y=y)
        ax.set_title(f'Box Plot: {col}')
```

```python
def plot_scatter(df, list_of_columns, y, figsize = (6,4)):
    for col in list_of_columns:
        fig = plt.figure(figsize=figsize)
        ax = sns.scatterplot(x=df[col], y=y)
        ax.set_title(f'Scatter Plot: {col}')
```

```python
def plot_swarm(df, list_of_columns, y, figsize = (6,4)):
    for col in list_of_columns:
        fig = plt.figure(figsize=figsize)
        ax = sns.swarmplot(data=df, x=df[col], y=y)
        ax.set_title(f'Swarm Plot: {col}')
```

```python
def plot_regplot(df, list_of_columns, y, figsize = (6,4)):
    for col in list_of_columns:
        fig = plt.figure(figsize=figsize)
        ax = sns.regplot(data=df, x=col, y=y)
        ax.set_title(f'Regression Plot: {col}')
```

```python
def plot_lmplot(df, list_of_columns, y, figsize = (6,4), legend=True):
    for col in list_of_columns:
        fig = plt.figure(figsize=figsize)
        ax = sns.lmplot(data=df, x=col, y=y)
```

```python
def plot_jointplot(df, list_of_columns, y, figsize = (6,4), legend=True):
    for col in list_of_columns:
        fig = plt.figure(figsize=figsize)
        ax = sns.jointplot(data=df, x=col, y=y, kind='reg')
```

```python
def plot_histograms(df, list_of_columns, ylabel=None, xlabel=None, bins=20,figsize=(6,4),vlines_list=None):
    for i, col in enumerate(list_of_columns):
        fig = plt.figure(figsize=figsize)
        ax = plt.hist(df[col], bins=bins, edgecolor='cornflowerblue')
        plt.vlines(x=df[col].mean(),
           ymin = 0,
           ymax = 450,
           color='firebrick',
           linestyle = 'solid',label='Mean')
        plt.vlines(x=df[col].median(),
           ymin = 0,
           ymax = 450,
           color='salmon',
           linestyle = 'dashed',label='Median')
        plt.vlines(x=df[col].quantile(q=0.25),
           ymin = 0,
           ymax = 450,
           color='lightsalmon',
           linestyle = 'dotted',label='25% Quantile')
        plt.vlines(x=df[col].quantile(q=0.75),
           ymin = 0,
           ymax = 450,
           color='lightsalmon',
           linestyle = 'dotted',label='75% Quantile')
        if vlines_list != None:
            for num in vlines_list:
                plt.vlines(x=num,
                           ymin = 0, ymax = 500, color='dimgrey',
                           linestyle = 'solid',label=f'{num}')
        plt.title(f'histogram: {col}',fontsize=18)
        plt.ylabel(ylabel, fontsize=10)
        plt.xlabel(xlabel, fontsize=10)
        plt.legend(fontsize=10);
        # plt.savefig('../plots/corr_heatmap.png');
```

```python
def subplot_box(df, list_of_columns, y, figsize=(10,8),
                ncols=2,nrows=0, sharex=True, sharey=True, figtitle=None):
    """plot histograms with subplots in matplotlib"""
    if nrows == 0:
        nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(figsize=figsize, nrows=nrows, ncols=ncols, 
                           sharex=sharex, sharey=sharey) # You'll want to specify your figsize
    fig.suptitle(figtitle, fontsize=16)
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns):         
        ax = sns.boxplot(x=df[column], y=y) #, ax=i)
        ax[i].set_title(f'Box Plot" {column}') 
        if (i >= (ncols*nrows)/2):
            ax[i].set_xlabel('column')  # only show x labels on bottom ax plots
        if (sharey==True) and (i % ncols == 0):
            ax[i].set_ylabel('y')  # only show y labels on left ax plots
    plt.tight_layout()
    # plt.savefig(f'../plots/{figtitle}.png');
```

```python
def plot_scatter_3d(x_column, y_column, hue, x_label, y_label, title):
    """plot single scatter plot using seaborn with hue on 3rd dimension"""
    plt.figure(figsize=(7,4))
    sns.scatterplot(x=x_column, y=y_column ,hue=hue)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.tight_layout()
    # plt.savefig(f'../plots/scatter_{title}.png')
```

**For the nominal values**, fill `'None'` if it represents 'absence of the particular feature' instead of a missing value.

```python
def fill_na_with_none(df, list_of_cols):
    """fill na/null values with 'None' where it represents absence of feature (not missing value)"""
    for i, column in enumerate(list_of_cols):
        df[column] = df[column].fillna('None')   # fill na with 'None'
        print(f'{column}   :   {df[column].unique()}')               # print to check the results with .unique
```

```python
# create functions for dropping columns and rows to be used throughout the project
# for columns - using list of columns to help drop multiple columns at a time (as part of feature selection)
# for rows - dropping one at a time to be more careful in losing number observations

def drop_col_train(df, list_columns_to_drop):
    """drop list of columns from train dataset"""
    print (f'train.shape before: {df.shape}\n')
    for col in list_columns_to_drop:        
        print (f'Dropped column : {col}')
        dropped_col_train.append(col)
        df.drop(columns = [col], inplace=True)
    print (f'\ntrain.shape after: {df.shape}\n')
    print ('\n')
        
def drop_col_test(df, list_columns_to_drop):
    """drop list of columns from test dataset"""
    print (f'test.shape before: {df.shape}\n')
    for col in list_columns_to_drop:        
        print (f'Dropped column : {col}')
        dropped_col_test.append(col)
        df.drop(columns = [col], inplace=True)
    print (f'\ntest.shape after: {df.shape}\n')
    print ('\n')

# defining functions to drop rows at the same time, this is for future use
    
def drop_row_train(df, index_row_to_drop):
    """drop row (one at a time) from train dataset"""
    print (f'train.shape before: {df.shape}\n')    
    print (f'Dropped row : {index_row_to_drop}')
    dropped_row_train.append(index_row_to_drop)
    df.drop(df.index[index_row_to_drop], inplace=True)
    print (f'\ntrain.shape after: {df.shape}\n')
    print ('\n')
    
def drop_row_test(df, index_row_to_drop):
    """drop row (one at a time) from train dataset"""
    print (f'test.shape before: {df.shape}\n')    
    print (f'Dropped row : {index_row_to_drop}')
    dropped_row_test.append(index_row_to_drop)
    df.drop(df.index[index_row_to_drop], inplace=True)
    print (f'\ntest.shape after: {df.shape}\n')
    print ('\n')
```

**For ordinal values** we need to assign a numerical value.

In this case we will use the following assumption where 3 represents average/typical/as expected:



|Score|Quality/Condition |Basement Finished Area|Basement Exposure|Garage Finish|Lot Shape|Lot Slope|Electrical|Paved Drive|  
|---|---|---|---|---|---|---|---|---|
|**5**|Excellent | | |
|**4**|Good|Good Living Quarters|Good Exposure| | | |Standard Circuit Breaker & Romex|
|**3**|Typical/Average|Average Living Quarters/ Average Rec Room|Average Exposure|Finished|Regular| |Average Fuse|Paved
|**2**|Fair|Below average living quarters|Minimum Exposure|Rough Finished|Slightly Irregular|Gentle Slope|Fair Fuse|Partial Pavement
|**1**|Poor|Low Quality or Unfinished|No Exposure|Unfinished|Moderately Irregular|Moderate Slope|Poor Fuse or Mix|Dirt/Gravel
|**0**|absence of feature|No Basement|No Basement|No Garage|Irregular|Severe Slope

- For **Fence**, only Good Privacy is value added to sales = 2, the rest are similar importance = 1
- For **Functional**, Range from 8 to 1 (8 = Typical, 7 = Minor Deductions 1....., 1 = Salvage only) 


```python
# check values for categorical data with ordinal values before assigning numerical value

for i, column in enumerate(na_columns_ordinal):
    print(column)
    print(train[column].unique())
```

```python
# for categorical data with ordinal values, we can assign a numerical value for analysis based on assumption above


ordinal_values = {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, np.nan : 0,
                  'Av':3, 'Mn':2, 'No':1, 'GLQ':4, 'ALQ':3, 'Rec':3, 
                  'BLQ':2, 'LwQ':1, 'Unf':1, 'Fin':3, 'RFn':2, 'Unf':1, 
                  'GdPrv':2, 'MnPrv':1, 'GdWo':1, 'MnWw':1,
                  'Reg':3, 'IR1':2, 'IR2':1, 'IR3':0,
                  'Gtl':2, 'Mod':1, 'Sev':0,
                  'SBrkr': 4, 'FuseA':3, 'FuseF':2, 'FuseP':1, 'Mix':1,
                  'Y':3,'P':2, 'N':1}   

ordinal_values_functional = {'Typ':8, 'Min1':7, 'Min2':6, 'Mod':5, 'Maj1':4, 'Maj2':3,'Sev':2,'Sal':1}   
```

```python
def assign_num_for_ordinal(df, list_of_cols):
    """assign numerical values for ordinal data types, fill 0 for absence of feature (not missing value)"""
    for i, column in enumerate(list_of_cols):        
        if column != 'functional':
            df[column] = df[column].map(ordinal_values)  # assign num to ordinal values
        else:
            df[column] = df[column].map(ordinal_values_functional) 
        print(column)
        print(df[column].unique())               # print to check the results with .unique
        print(df[column].dtype)
        print('\n')
```

```python
# use function to assign numbers for ordinal variables with null values

assign_num_for_ordinal(train, na_columns_ordinal)

assign_num_for_ordinal(test, na_columns_ordinal)
```

## 2.3 EDA, Feature Selection, and Feature Engineering



### <span style='color:darkblue'>Drop features with one dominant value (>95% of data on one value)</span>  

- When a feature has majority of data on a single value, this indicates a "standard" feature, with the minority of houses having "special variations".  Usually when the standard feature is in almost every house, it will be regardless of the `saleprice` and hence would not a good predictor for the majority, while the 'special variations are too specific and less likely be able to generalize to new data.
- We assume a feature with >95% of data on one value as having a dominant value, which will be dropped after validating with boxplots or swarmplots against `'saleprice'`.

**Example:** For Roof Material `'Roof Matl'` 2025 out of 2051 observations are standard composite shingle `'CompShg'`, and corresponds to `saleprice` ranging from min to max of house prices (close to $0 to over $600,000).  Therefore, `'Roof Matl'` will not be a good predictor of `saleprice`.

We will apply the same analysis on the rest of the 'dominant' features.

```python
train.groupby(train['roof_matl'])['roof_matl'].count().values
```

```python
sns.scatterplot(x=train['roof_matl'],y=train.saleprice)
```

```python
# Create a list of features with dominant value

dominant_list = []
for col in train.columns:
    if (col != 'kitchen_abvgr'                  # keep 'kitchen above ground' to calculate rooms in later stage
       ) and (col != 'three_ssn_porch'):         # keep 'three season porch' to calculate sqft in later stage
        gb_array = (train.groupby(train[col])[col].count().values)
        for i in gb_array:
            if i > (2051*.95):
                dominant_list.append(col)

dominant_list
```

From the swarm plots of all the dominant features list, the dominant values ranges across different sale prices (low to high), while the non-dominant values scattered within the range, making these features less predictive of `'saleprice'`.  Hence will drop all these features.
