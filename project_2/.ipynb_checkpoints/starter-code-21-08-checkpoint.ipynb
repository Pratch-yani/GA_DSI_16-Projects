**For the nominal values**, fill `'None'` if it represents 'absence of the particular feature' instead of a missing value.

```python
def fill_na_with_none(df, list_of_cols):
    """fill na/null values with 'None' where it represents absence of feature (not missing value)"""
    for i, column in enumerate(list_of_cols):
        df[column] = df[column].fillna('None')   # fill na with 'None'
        print(f'{column}   :   {df[column].unique()}')               # print to check the results with .unique
```

```python
# fill NaN with 'None' for both train and test datasets

fill_na_with_none(train,na_columns_nominal)

fill_na_with_none(test,na_columns_nominal)
```

**For ordinal values** we need to assign a numerical value.

In this case we will use the following assumption:
(Note: the list below covers all features with ordinal values that requires assignment to numerical)


- 5 = Excellent 
- 4 = Good, Good Living Quarters, Standard Circuit Breaker & Romex
- 3 = Typical, Average, Average Living Quarters, Average Rec Room, Finished, Regular, Average Fuse
- 2 = Fair, Below average living quarters, Rough Finished, Slightly Irregular, Gentle Slope, Fair Fuse
- 1 = Poor, Low Quality, Unfinished, Moderately Irregular, Moderate Slope, Poor Fuse or Mix
- 0 = NaN, absence of that feature (i.e. garage, basement, etc), Irregular, Severe Slope
- For Fence, only Good Privacy is value added to sales = 2, the rest are similar importance = 1
- For Functional, Range from 8 to 1 (8 = Typical, 7 = Minor Deductions 1....., 1 = Salvage only) 


```python
# check values for categorical data with ordinal values before assigning numerical value

for i, column in enumerate(na_columns_ordinal):
    print(column)
    print(train[column].unique())
```

```python
# for categorical data with ordinal values, we can assign a numerical value for analysis based on assumption above


ordinal_values = {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, np.nan : 0,
                  'Av':3, 'Mn':2, 'No':1, 'GLQ':4, 'ALQ':3, 'Rec':3, 
                  'BLQ':2, 'LwQ':1, 'Unf':1, 'Fin':3, 'RFn':2, 'Unf':1, 
                  'GdPrv':2, 'MnPrv':1, 'GdWo':1, 'MnWw':1,
                  'Reg':3, 'IR1':2, 'IR2':1, 'IR3':0,
                  'Gtl':2, 'Mod':1, 'Sev':0,
                  'SBrkr': 4, 'FuseA':3, 'FuseF':2, 'FuseP':1, 'Mix':1}   

ordinal_values_functional = {'Typ':8, 'Min1':7, 'Min2':6, 'Mod':5, 'Maj1':4, 'Maj2':3,'Sev':2,'Sal':1}   
```

```python
def assign_num_for_ordinal(df, list_of_cols):
    """assign numerical values for ordinal data types, fill 0 for absence of feature (not missing value)"""
    for i, column in enumerate(list_of_cols):        
        if column != 'functional':
            df[column] = df[column].map(ordinal_values)  # assign num to ordinal values
        else:
            df[column] = df[column].map(ordinal_values_functional) 
        print(column)
        print(df[column].unique())               # print to check the results with .unique
        print(df[column].dtype)
        print('\n')
```

```python
# use function to assign numbers for ordinal variables

assign_num_for_ordinal(train, na_columns_ordinal)

assign_num_for_ordinal(test, na_columns_ordinal)
```

```python
def plot_heatmap(title, corr, figsize = (12,12),cmap="RdBu_r"):
    fig = plt.figure(figsize=figsize)
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True
    ax = sns.heatmap(corr, mask=mask, annot=True,cmap=cmap)
    ax.set_title(f'Correlation Heatmap: {title}')

# plt.savefig('../plots/corr_heatmap.png');
```

```python
def plot_scatter(df, x, y, label, figsize = (6,4)):
    fig = plt.figure(figsize=figsize)
    ax = sns.scatterplot(data=df, x=x, y=y)
    ax.set_title(f'Scatter Plot: {label}')

# plt.savefig('../plots/corr_heatmap.png');
```

```python
def plot_regplot(df, x, y, label, figsize = (6,4)):
    fig = plt.figure(figsize=figsize)
    ax = sns.regplot(data=df, x=x, y=y)
    ax.set_title(f'Regression Plot: {label}')

# plt.savefig('../plots/corr_heatmap.png');
```

```python
grg_yr_corr = train[['garage_yr_blt', 'year_built','year_remod/add','saleprice']].corr()
plot_heatmap('Garage Year Built', grg_yr_corr, figsize=(5,5),cmap='Blues')
```

```python
def plot_histograms(df, list_of_columns, ylabel=None, xlabel=None, bins=20,figsize=(6,4)):
    for i, col in enumerate(list_of_columns):
        fig = plt.figure(figsize=figsize)
        ax = plt.hist(df[col], bins=20)
        plt.vlines(x=df[col].mean(),
           ymin = 0,
           ymax = 450,
           color='red',
           linestyle = '--',label='Mean')
        plt.vlines(x=df[col].median(),
           ymin = 0,
           ymax = 450,
           color='orange',
           linestyle = '--',label='Median')
        plt.title(f'histogram: {col}',fontsize=18)
        plt.ylabel(ylabel, fontsize=10)
        plt.xlabel(xlabel, fontsize=10)
        plt.legend(fontsize=10);
        # plt.savefig('../plots/corr_heatmap.png');
```

### Drop features with one dominant value (>95% of data on one value)  

- When a feature has majority of data on a single value, this indicates a "standard" for that feature, with the minority of houses having "special variations".  In this case, this feature would not a good predictor for as majority of the data would have standard feature.
- We assume a feature with >95% of data on one value as having a dominant value, which will be dropped after validating with boxplots or swarmplots against `'saleprice'`.

```python
def compare_histograms(df, imputed_column, original_column, missingness_column, x_label, y_label = 'Frequency'):
    fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (16,9))

    # Set axes of histograms.
    mode = stats.mode(df[imputed_column])
    rnge = max(df[original_column]) - min(df[original_column])
    xmin = min(df[original_column]) - 0.02 * rnge
    xmax = max(df[original_column]) + 0.02 * rnge
    ymax = 1.3 * (mode[1][0] + df[df[original_column] == mode[0][0]].shape[0])

    ax0.set_xlim(xmin, xmax)
    ax0.set_ylim(0, ymax)
    ax1.set_xlim(xmin, xmax)
    ax1.set_ylim(0, ymax)

    # Set top labels.
    ax0.set_title('Real Histogram', position = (0,1), ha = 'left', fontsize = 25)
    ax0.set_xlabel(x_label, position = (0,0), ha = 'left', fontsize = 25, color = 'grey', alpha = 0.85)
    ax0.set_ylabel(y_label, position = (0,1), ha = 'right', va = 'top', fontsize = 25, rotation = 0, color = 'grey', alpha = 0.85)
    ax0.set_xticks([])
    ax0.set_yticks([])

    # Generate top histogram.
    ax0.hist(df[original_column], bins = 15, color = '#185fad', alpha = 0.75, label = '')
    ax0.axvline(np.mean(df[original_column]), color = '#185fad', lw = 5, label = 'True Mean')
    ax0.legend(prop={'size': 15}, loc = 1)

    # Set bottom labels.
    ax1.set_title('Observed + Imputed Histogram', position = (0,1), ha = 'left', fontsize = 25)
    ax1.set_xlabel(x_label, position = (0,0), ha = 'left', fontsize = 25, color = 'grey', alpha = 0.85)
    ax1.set_ylabel(y_label, position = (0,1), ha = 'right', va = 'top', fontsize = 25, rotation = 0, color = 'grey', alpha = 0.85)

    # Generate bottom histogram.
    ax1.hist([df[imputed_column][df[missingness_column] == 0], df[imputed_column][df[missingness_column] == 1]], bins = 15, color = ['#185fad','orange'], alpha = 0.75, label = '', stacked = True)
    ax1.axvline(np.mean(df[original_column]), color = '#185fad', lw = 5, label = 'True Mean')
    ax1.axvline(np.mean(df[original_column][df[missingness_column] == 0]), color = 'grey', alpha = 0.5, lw = 5, label = 'Observed Mean')
    ax1.axvline(np.mean(df[imputed_column]), color = 'orange', lw = 5, label = 'Observed and Imputed Mean')
    ax1.legend(prop={'size': 15}, loc = 1)

    plt.show();
```


*Key Findings & Actions to be taken*
- **<span style='color:royalblue'> Drop Idenfification Numbers </span>** (`'Id'`, `'PID'`) are for the purpose of record-keeping and identification, and are less likely to impact the sales price of a house.

- Some `NaN` can be corrected by filling as `'None'`.  Based on the [data description](http://jse.amstat.org/v19n3/decock/DataDocumentation.txt), the use of 'NA' represent cases where the property does not have that feature, hence no score, but it does not equate to a 'missing/null value'.  For example, for `'Alley'`, NA = 'No Alley Access', hence can be filled with 'None' instead of a `NaN` / null value.

- **<span style='color:royalblue'> Drop features where most observations fall in 1 category</span>** 
  - `'Utilities'`, `'Roof Matl'`
